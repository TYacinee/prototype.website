### 3. Extracting the data

Then we can extract the data. Here is a small sample of examples of data that we can get thanks to the API key.

```python
# Extraction of the data.

response = requests.get(BASE_URL, headers=HEADERS, params=params)

if response.status_code != 200:
    print(f"Error {response.status_code} : {response.text}") # This allow us to see if our API key works and allow us to access to the website.
else:
    data = response.json()
    replays = data.get("list", [])
    
    
    # Examples of data that we can get.
    
    all_replays = []
    for replay in replays:
        all_replays.append({
            "id": replay.get("id"),
            "title": replay.get("title"),
            "map": replay.get("map_code"),
            "date": replay.get("date"),
            "duration": replay.get("duration"),
            "uploader": replay.get("uploader", {}).get("name"),
            "playlist": replay.get("playlist_id"),
            "blue_score": replay.get("blue", {}).get("score"),
            "orange_score": replay.get("orange", {}).get("score"),
        })
```
